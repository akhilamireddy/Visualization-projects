```python
import numpy as np #importing libraries
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/Telegram Desktop/Viz_Project/Canada.csv") #importing .CSV file
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__)
```

```python
dp=df.dropna(how="all",axis=1) #drop columns from the DataFrame df that contain all missing values (NaN) along the specified axis.
dp
```

```python
df_can=dp.dropna(how="all",axis=0) #drop columns from the DataFrame df that contain all missing values (NaN) along the specified axis.
df_can
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True) #used to drop multiple columns from the DataFrame df_can.
df_can
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

# let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
#checking wheather all coloumns are strings or not
all(isinstance(column, str) for column in df_can.columns)
```

```python
#converting all coloumns to strings
df_can.columns = list(map(str, df_can.columns))

# let's check the column labels types now
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.set_index('Country', inplace=True)

# let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1)

# let's view the first five elements and see how the dataframe was changed
df_can.Total
```

# i. Create an area plot for top 6 immigrant countries from 1990 to 2013

```python
years_1 = list(map(str, range(1990, 2014))) #creating a list that contains years from 1990-2014
years_1
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True) #is used to sort the DataFrame 'df_can' based on the values in the column 'Total' in descending order.

# get the top 6 entries
df_top6 = df_can.head(6)

# transpose the dataframe
df_top6 = df_top6[years_1].transpose() 

df_top6
```

```python
df_top6.index = df_top6.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top6.plot(kind='area', stacked=False,figsize=(20, 10)) # pass a tuple (x, y) size

plt.title('Immigration Trend of Top 6 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

# ii. Create and year-wise immigrant bar chart from India to Canada during the period of 1980 to 2013

```python
years_2=list(map(str, range(1980, 2014))) #creating a list that contains years from 1980-2014
years_2
```

```python
df_india = df_can.loc['India', years_2] #is used to select a specific row and columns from the DataFrame 'df_can' and store it in a new DataFrame called 'df_india'.
df_india
```

```python
df_india.plot(kind='bar', figsize=(13, 6)) #plotting bar plot

plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('India immigrants to Canada from 1980 to 2013') # add title to the plot

plt.show()
```

# iii) Create a boxplot for Indian, Phillipins and China immigrants.

```python
df_CI = df_can.loc[['China','India','Philippines'],years_2].transpose() #is used to select specific rows and columns from the DataFrame 'df_can', transpose the resulting DataFrame, and store it in a new DataFrame called 'df_CI'. 
df_CI
```

```python
df_CI.plot(kind='box',
          figsize=(6,4)) #plotting box plot
plt.title('Box plots of Immigrants from China and India (1980 - 2013)')
plt.xlabel('Number of Immigrants')
plt.show()
```

# iv) Show the total no. of immigrants from India and France countries using Area Chart and Pie chart.

```python
df_IF = df_can.loc[['India', 'France'],'Total'] #used to select specific rows and a specific column from the DataFrame 'df_can' and store it in a new DataFrame called 'df_IF'.
df_IF
```

```python
df_IF.plot(kind='pie',
                            figsize=(3, 4),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                           )

plt.title('Immigration to Canada from India and France [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```
